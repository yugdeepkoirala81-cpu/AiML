```python
import numpy as np
```

# concatenation in numpy

```python
ar1=np.array([1,2,3,4])
ar2=np.array([2,2,2,2])
ar3=np.concatenate((ar1,ar2))
print(ar3) # axis:1 concatenation
```

# concatenation in 2d array in axis=0

```python
a1=np.array([[1,2,3,4],[5,3,4,5]])
a2=np.array([[5,6,7,8],[6,7,8,9]])
a3=np.concatenate((a1,a2), axis=0)
print(a3)  #axis=0 concatenation
```

# concatenation in 2d array in axis=1

```python
a1=np.array([[1,2,3,4],[5,3,4,5]])
a2=np.array([[5,6,7,8],[6,7,8,9]])
a3=np.concatenate((a1,a2), axis=1)
print(a3)  
```

# concatenating an array to either of the axis

```python
a1=np.array([[1,2,3,4],[5,3,4,5]])
a2=np.array([[5,6,7,8],[6,7,8,9]])
a4=np.concatenate((a1,a2),axis=1)    # for adding value to the side of the array
print(a4)
a5=np.concatenate((a1,a2),axis=0)   # for adding value below the existing array
print(a5)
```

# vstack and hstack

```python
a3= np.vstack((a1,a2))
print(a3)
a7=np.hstack((a1,a2))
print(a7)
```
